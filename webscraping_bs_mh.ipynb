```python
from bs4 import BeautifulSoup as BS
```

```python
soup = BS(response.text)
```

```python
print(soup.prettify())
```

```python
soup.find('h2', class_='title is-5').text
```

```python
soup.find('p', class_='location').text
```

```python
soup.find('time').text
```

```python
titles = [t.get_text(strip=True) for t in soup.find_all('h2')]
titles
```

```python
companies = [t.get_text(strip=True) for t in soup.find_all('h3')]
print(type(companies))
companies
```

```python
locations = [t.get_text(strip=True) for t in soup.findAll('p', class_='location')]
print(type(locations))
locations
```

```python
postdate = [t.get_text(strip=True) for t in soup.find_all('time')]
print(type(postdate))
postdate
```

```python
import pandas as pd
```

```python
jobs = pd.DataFrame({
    'Title': titles,
    'Company': companies,
    'Location': locations,
    'Date Posted': postdate
})
jobs.head()
```

```python
job_descriptions = []

for each_url in apply:
    each_url = requests.get(each_url)
    soup_url = BS(each_url.text)
    text = soup_url.findAll('p')[1].text
    job_descriptions.append(text)

job_descriptions
```

```python
def get_desc(singleURL):
    request = requests.get(singleURL)
    soup = BS(request.text)
    return soup.findAll('p')[1].get_text(strip=True)
```

```python
desc_list = jobs['Apply Link'].apply(get_desc)
```

```python
jobs['Description'] = desc_list
```
